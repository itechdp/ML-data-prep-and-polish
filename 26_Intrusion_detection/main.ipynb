## Data Preprocessing
- Standardization of Data
- Splitting into Features and Labels 
- Train Test Split

### Standardization of data

```python
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df)
df_normalized = pd.DataFrame(df_normalized,columns=df.columns)
```

### Splitting into Features and Label

```python
df_features = df_normalized.drop('Number of Barriers',axis=1)
df_label = df_normalized[['Number of Barriers']]
```

### Train and Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df_features,df_label,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```
