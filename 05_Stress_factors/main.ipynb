# Stress Factor Dataset

### Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

### Importing Dataset

```python
df = pd.read_csv('Student Stress Factors.csv')
df.head()
```

### Information about Attributes
- As per the below table, we could analyse that, we need to perform Timestamp encoding on Timestamp column. Hence it is already numerical data
we do not require to perform encoding on other data columns.

```python
df_info = df.info()
df_info
```

### Describing Dataset
- Just because it is rating from 0 to 5, so we will not be preforming Removing outliers from this dataset because all datapoints ranges between 0 to 5. If we will get any data point > 5 then we will remove that. Hence we dont have any data point so we won't be removing outliers.

```python
df_describe = df.describe()
df_describe
```

### Handling Timestamp data 

```python
# Converting it to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['year'] = df['Timestamp'].dt.year
df['month'] = df['Timestamp'].dt.month
df['day'] = df['Timestamp'].dt.day
df['hour'] = df['Timestamp'].dt.hour
df['minute'] = df['Timestamp'].dt.minute
df['second'] = df['Timestamp'].dt.second
print(df.iloc[:,7:13]) # Acessing Columns Using iloc[rows:column]
print(df.loc[:,'year':'second']) # Acessing Columns using loc[row:column name]
```

### Splitting into Features and labels

```python
df_features = df.drop(['stress levels','Timestamp'],axis=1)
df_label = df[['stress levels']]
```

### Plotting Hist Plot

```python
for cols in df_features.columns:
    plt.hist(x=df_features[cols])
    plt.xlabel(cols)
    plt.ylabel("Count")
    plt.show()
```

### Plotting bar chart 

```python
plt.bar(df_features.T.columns,df_features.T.iloc[0])
plt.xlabel('Instance')
plt.ylabel("Value count")
```

### Plotting Scatter plot

```python
plt.scatter(x=df_features['Sleep Quality'],y=df_label['stress levels'])
plt.xlabel("Sleep Quality")
plt.ylabel('Stress levels')
```
