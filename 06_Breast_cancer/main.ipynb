# Breast Cancer Dataset

### Importing Packages

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
df = pd.read_csv('wdbc.data')
df.head()
```

### Assigning Columns

```python
dataset_cols = ['ID', 'Diagnosis', 'radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1', 'compactness1', 'concavity1', 'concave_points1', 'symmetry1', 'fractcal_dimension1', 'radius2', 'texture2', 'perimeter2', 'area2', 'smoothness2', 'compactness2', 'concavity2', 'concave_points2', 'symmetry2', 'fractcal_dimension2', 'radius3', 'texture3', 'perimeter3', 'area3', 'smoothness3', 'compactness3', 'concavity3', 'concave_points3', 'symmetry3', 'fractcal_dimension3']
df.columns = dataset_cols
```

### Information of DataFrame

```python
df_info = df.info()
print(df_info) 
```

### Describe DataFrame

```python
df_describe = df.describe()
df_describe
```

### Splitting into Featues and Label

```python
df_features = df.drop('Diagnosis',axis=1)
df_label = df[['Diagnosis']]
```

### Information about Features
- Just because of we are having each and every column with numerical Dtype we will be applying Normalization of data.

```python
df_features_info = df_features.info()
df_features_info
```

### Information about Label

```python
df_label_info = df_label.info()
df_label_info
```

### Checking for N-unique value for Label

```python
df_label.nunique()
```

### Checking values for N-unique values in Label

```python
df_label.value_counts()
```

### Encoding of Label

```python
label_encoder = LabelEncoder()
df_label_encoded = label_encoder.fit_transform(df_label['Diagnosis'])
df_label_encoded = pd.DataFrame(df_label_encoded,columns=df_label.columns)
df_label_encoded.value_counts()
```

### Checking for Correlations

```python
df_correaltions = pd.concat([df_features,df_label_encoded],axis=1).corr()
plt.figure(figsize=(25,20))
df_correaltions_Plot = sns.heatmap(df_correaltions,annot=True)
plt.title("Correlation of Features & Label")
```

### Checking for the outliers

```python
# plt.figure(figsize=(20,5))
for cols in df_features.columns:
    df_features[[cols]].boxplot()
    plt.title("Box plot of Feature column")
    plt.ylabel("Value counts")
    plt.tight_layout()
    plt.show()
```

### Removing outliers

```python
Q1 = df_features.quantile(0.25)
Q3 = df_features.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_features_no_outliers = df_features[(df_features >= lower_bound) & (df_features <= upper_bound)]
for cols in df_features_no_outliers.columns:
    df_features_no_outliers[[cols]].boxplot()
    plt.title("Box plot of Feature column")
    plt.ylabel("Value counts")
    plt.tight_layout()
    plt.show()
```

### Normalizing the data

```python
scaler = MinMaxScaler()
df_features_scaled = scaler.fit_transform(df_features_no_outliers)
df_features_scaled = pd.DataFrame(df_features_scaled,columns=df_features.columns)
```

### Final Feature and label
- Now you can use `df_final_features` and `df_final_label` as a processed feature and label. Now machine learning could be applied on these feature and label variables.

```python
df_final_features = df_features_scaled.copy()
df_final_label = df_label_encoded.copy()
```
